# **PREDICTING STUDENT MARKS**

## IMPORTING MODELS

## READING DATA FILE

```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
model=LinearRegression()
```

```python
df=pd.read_csv("student_marks_dataset.csv")
df.head()
```

```python
x= df[["hours_studied", "attendance", "previous_score"]]
y = df["final_marks"]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
model.fit(x_train,y_train)
```

```python
pred=model.predict(x_test)
```

```python
score=model.score(x_test,y_test)
print(score)
```

```python
from matplotlib import pyplot as plt
plt.scatter(y_test,pred)
plt.xlabel("Actual Marks")
plt.ylabel("Predicted Marks")
plt.title(" Actual vs Predicted Marks")
plt.plot(y_train,(y_train),"#F8766D")
```

```python
hours = float(input("Enter hours studied: "))
attendance = float(input("Enter attendance (%): "))
previous = float(input("Enter previous score: "))
data = [[hours, attendance, previous]]
predicted_marks = model.predict(data)
predicted_marks = max(0, min(predicted_marks, 100))
print("Predicted final marks:", predicted_marks[0])
```
